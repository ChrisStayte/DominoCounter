Taking annotated images from my domino image library and augmenting them

```python
# Import images and labels and augment them using albumentations so I can train a yolo v5 model

import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import albumentations as A
from albumentations.pytorch.transforms import ToTensorV2
from tqdm import tqdm

# Import images and labels
images_path = '../data/images/'
labels_path = '../data/labels/'

images = os.listdir(images_path)
labels = os.listdir(labels_path)

# Sort images and labels
images.sort()
labels.sort()

# Create a dataframe with the images and labels
df = pd.DataFrame({'images': images, 'labels': labels})

# Create a function to get the bounding boxes
def get_bboxes(path):
    with open(path, 'r') as f:
        lines = f.readlines()
        bboxes = []
        for line in lines:
            bbox = line.split()
            bboxes.append(bbox)
    return bboxes

# Create a function to get the class names
def get_class_names(path):
    with open(path, 'r') as f:
        lines = f.readlines()
        class_names = []
        for line in lines:
            class_name = line.split()[0]
            class_names.append(class_name)
    return class_names

# Create a function to get the class ids
def get_class_ids(path):
    with open(path, 'r') as f:
        lines = f.readlines()
        class_ids = []
        for line in lines:
            class_id = line.split()[1]
            class_ids.append(class_id)
    return class_ids



```

```python
# import libraries used for data augmentation using albumentations
import albumentations as A
from albumentations.pytorch import ToTensorV2
from torchvision.datasets import ImageFolder



# define the transformations to be applied on the training data
# define the paths to the training, validation and test directories
train_dir = "/Users/chrisstayte/Desktop/Doubletwelvedominoes/train/images"
val_dir = "/Users/chrisstayte/Desktop/Doubletwelvedominoes/valid/images"
test_dir = "/Users/chrisstayte/Desktop/Doubletwelvedominoes/test/images"

train_transforms = A.Compose([
    A.Resize(height=256, width=256),
    A.RandomCrop(height=224, width=224),
    A.Rotate(limit=35, p=1.0),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.1),
    A.RandomBrightnessContrast(p=0.2),
    A.Normalize(
        mean=[0.0, 0.0, 0.0],
        std=[1.0, 1.0, 1.0],
        max_pixel_value=255.0,
    ),
    ToTensorV2(),
])

# define the transformations to be applied on the validation data
val_transforms = A.Compose([
    A.Resize(height=256, width=256),
    A.CenterCrop(height=224, width=224),
    A.Normalize(
        mean=[0.0, 0.0, 0.0],
        std=[1.0, 1.0, 1.0],
        max_pixel_value=255.0,
    ),
    ToTensorV2(),
])

# define the transformations to be applied on the test data
test_transforms = A.Compose([
    A.Resize(height=256, width=256),
    A.CenterCrop(height=224, width=224),
    A.Normalize(
        mean=[0.0, 0.0, 0.0],
        std=[1.0, 1.0, 1.0],
        max_pixel_value=255.0,
    ),
    ToTensorV2(),
])

# setup the training, validation and test datasets
train_dataset = ImageFolder(
    root=train_dir,
    transform=train_transforms
)

val_dataset = ImageFolder( 
    root=val_dir,
    transform=val_transforms
)

test_dataset = ImageFolder(
    root=test_dir,
    transform=test_transforms
)



```
